```python
from transformers import pipeline
```

```python
import pandas as pd
```

```python
classifier = pipeline("sentiment-analysis", model="HooshvareLab/bert-fa-base-uncased-sentiment-snappfood")
```

```python
df = pd.read_csv("comment_10.csv")
```

```python
comments = df["cleaned_comment"]
```

```python
def analyze_sentiment(text):
    result = classifier(text)[0]
    return result['label']
```

```python
data = {"comment": ["خیلی محصول خوبی بود!", "بدون کیفیت و ضعیف", "معمولی بود"]}
df = pd.DataFrame(data)
```

```python
df['emotion'] = df['comment'].apply(analyze_sentiment)
df
```

```python
df['emotion'] = df['cleaned_comment'].apply(analyze_sentiment)
df
```
